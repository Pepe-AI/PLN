```python
def preprocess(text):
    # Tokeniza el texto
    words = word_tokenize(text)

    # Cambia las palabras a minúsculas
    words = [word.lower() for word in words]

    # Lematiza las palabras
    words = [lemmatizer.lemmatize(word) for word in words]

    # Une las palabras de nuevo en un solo string
    text = ' '.join(words)
    return text

files = ['cadiz.txt', 'bailén.txt', 'el_19_de_marzo_y_el_2_de_mayo.txt', 'el_abuelo.txt', 'el_amigo_manso.txt', 'el_caballero_encantado.txt', 'el_doctor_centeno_1.txt', 'el_doctor_centeno_2.txt',
         'el_equipaje_del_rey_jose.txt', 'la_batalla_de_los_arapiles.txt', 'la_carrera_de_san_jeronimo.txt', 'la_corte_de_carlos_iv.txt', 'la_de_bringas.txt', 'la_desheredada.txt', 'trafalgar.txt', 'un_fracaso_mas_y_algunos_frailes_menos.txt',
         'argentina.txt', 'arroz_y_tartana.txt', 'cañas_y_barro.txt', 'cuentos_valencianos.txt', 'el_intruso.txt', 'el_jardin_de_la_capilla_exploratoria.txt', 'entre_naranjos.txt', 'flor_de_mayo.txt',
         'la_barraca.txt', 'la_bodega.txt', 'la_catedral.txt', 'la_condenada.txt', 'la_horda.txt', 'la_novela_de_hoy.txt', 'los_argonautas.txt', 'Los_enemigos_de_la mujer.txt']


# Inicializa una lista para almacenar el texto de los archivos
texts = []

# Itera sobre los archivos y lee su contenido
for file in files:
    with open(file, 'r', encoding='utf-8') as f:
        texts.append(f.read())

# Descarga los recursos necesarios de nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Define el lematizador y las stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('spanish'))



# Itera sobre los archivos y preprocesa su contenido
for i in range(len(texts)):
    texts[i] = preprocess(texts[i])

print(texts[1:2])
```

```python

# Función para contar las oraciones en un texto
def count_sentences(text):
    sentences = nltk.tokenize.sent_tokenize(text)
    return len(sentences)

def count_words_in_file(text):
    words = nltk.tokenize.word_tokenize(text)
    return len(words)

def count_unique_words_in_file(text):
    words = nltk.tokenize.word_tokenize(text)
    unique_words = set(words)  # crea un conjunto, que elimina las palabras duplicadas
    return len(unique_words)

def count_stopwords_in_file(text):
    words = nltk.word_tokenize(text)
    stop_words = set(stopwords.words('spanish'))
    stopwords_in_text = [word for word in words if word in stop_words]
    return len(stopwords_in_text)

def count_punctuation_in_file(text):
    punctuation_chars = set(string.punctuation)
    punctuation_in_text = [char for char in text if char in punctuation_chars]
    return len(punctuation_in_text)


def count_syllables_in_file(text):
    dic = pyphen.Pyphen(lang='es')
    words = nltk.word_tokenize(text)
    syllables = [dic.inserted(word) for word in words]
    num_syllables = sum([len(word.split('-')) for word in syllables])
    return num_syllables

files = ['cadiz.txt', 'bailén.txt', 'el_19_de_marzo_y_el_2_de_mayo.txt', 'el_abuelo.txt', 'el_amigo_manso.txt', 'el_caballero_encantado.txt', 'el_doctor_centeno_1.txt', 'el_doctor_centeno_2.txt',
         'el_equipaje_del_rey_jose.txt', 'la_batalla_de_los_arapiles.txt', 'la_carrera_de_san_jeronimo.txt', 'la_corte_de_carlos_iv.txt', 'la_de_bringas.txt', 'la_desheredada.txt', 'trafalgar.txt', 'un_fracaso_mas_y_algunos_frailes_menos.txt',
         'argentina.txt', 'arroz_y_tartana.txt', 'cañas_y_barro.txt', 'cuentos_valencianos.txt', 'el_intruso.txt', 'el_jardin_de_la_capilla_exploratoria.txt', 'entre_naranjos.txt', 'flor_de_mayo.txt',
         'la_barraca.txt', 'la_bodega.txt', 'la_catedral.txt', 'la_condenada.txt', 'la_horda.txt', 'la_novela_de_hoy.txt', 'los_argonautas.txt', 'Los_enemigos_de_la mujer.txt']

# Asigna los autores a los libros (asumiendo que los primeros 16 libros son del autor1 y los siguientes 16 del autor2)
authors = ['Benito Pérez Galdós']* 16  + ['Blasco Ibañez Vicente']*16

# Inicializa una lista para almacenar el texto de los archivos
texts = []

# Itera sobre los archivos y lee su contenido
for file in files:
    with open(file, 'r', encoding='utf-8') as f:
        texts.append(f.read())


# Crea un DataFrame con el índice siendo el nombre del archivo y la columna "text" siendo el texto del archivo
data = pd.DataFrame({'author': authors, 'text': texts, 'nombre': files})


# Agrega una columna al DataFrame con el número de oraciones para cada texto
data['num_sentences'] = data['text'].apply(count_sentences)
data = data.drop('text', axis=1)


# Cuenta las palabras en cada archivo y almacena los resultados en una nueva columna del DataFrame
data['num_words'] = [count_words_in_file(file) for file in texts]


# Cuenta las palabras únicas en cada archivo y almacena los resultados en una nueva columna del DataFrame
data['num_unique_words'] = [count_unique_words_in_file(file) for file in texts]


# Cuenta las stopwords en cada archivo y almacena los resultados en una nueva columna del DataFrame
data['num_stopwords'] = [count_stopwords_in_file(file) for file in texts]


# Cuenta los signos de puntuación en cada archivo y almacena los resultados en una nueva columna del DataFrame
data['num_punctuation'] = [count_punctuation_in_file(file) for file in texts]


# Calcula el ratio y almacénalo en una nueva columna 'punctuation_to_sentence_ratio'
data['punctuation_to_sentence_ratio'] = data['num_punctuation'] / data['num_sentences']


# Para evitar la división por cero, puedes reemplazar los valores infinitos por NaN
data['punctuation_to_sentence_ratio'].replace([np.inf, -np.inf], np.nan, inplace=True)


# Calcula el ratio y almacénalo en una nueva columna 'unique_to_total_words_ratio'
data['unique_to_total_words_ratio'] = data['num_unique_words'] / data['num_words']


# Para evitar la división por cero, puedes reemplazar los valores infinitos por NaN
data['unique_to_total_words_ratio'].replace([np.inf, -np.inf], np.nan, inplace=True)


# Calcula el ratio y almacénalo en una nueva columna 'stopwords_to_total_words_ratio'
data['stopwords_to_total_words_ratio'] = data['num_stopwords'] / data['num_words']


# Para evitar la división por cero, puedes reemplazar los valores infinitos por NaN
data['stopwords_to_total_words_ratio'].replace([np.inf, -np.inf], np.nan, inplace=True)


# Cuenta las sílabas en cada archivo y almacena los resultados en una nueva columna del DataFrame
data['num_syllables'] = [count_syllables_in_file(file) for file in texts]


# Calcula el FRE y almacénalo en una nueva columna 'FRE'
data['FRE'] = 206.835 - 1.015 * (data['num_words'] / data['num_sentences']) - 84.6 * (data['num_syllables'] / data['num_words'])


# Calcula el FKGL y almacénalo en una nueva columna 'FKGL'
data['FKGL'] = 0.39 * (data['num_words'] / data['num_sentences']) + 11.8 * (data['num_syllables'] / data['num_words']) - 15.59
print(data)

data.to_csv('dataset.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics

df = pd.read_csv('dataset.csv')
df = df.drop('nombre', axis=1)

# Supongamos que 'columna_etiqueta' es el nombre de tu columna de clasificación
y = df['author']


# Elimina la columna de la etiqueta para obtener solo las características
X = df.drop('author', axis=1)


# Divide los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)


# Crea un clasificador Gaussian Naive Bayes y entrena con los datos
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Hacer predicciones
y_pred = gnb.predict(X_test)

# Evalúa el modelo
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
